#### To count the number of outliers in each numeric variable:

```python
Q1 = dfnumeric.quantile(0.25)
Q3 = dfnumeric.quantile(0.75)

# Rule to identify outliers
rule = ((dfnumeric < (Q1 - 1.5 * (Q3 - Q1))) | (dfnumeric > (Q3 + 1.5 * (Q3 - Q1))))

# Count the number of outliers
rule.sum()
```

```python
# Group the crop data in groups of 100 rows
groupCrops = [dfnumeric[i:i+100] for i in range(0, len(dfnumeric), 100)]

# Find outliers using the IQR method for each group
outliers = pd.DataFrame()
for crop in groupCrops:
    q1 = crop.quantile(0.25)
    q3 = crop.quantile(0.75)
    iqr = q3 - q1
    individualCrop_outliers = crop[((crop < (q1 - 1.5 * iqr)) | (crop > (q3 + 1.5 * iqr))).any(axis=1)]
    outliers = pd.concat([outliers, individualCrop_outliers])

# Print the outliers 
outliers
```

- This code shows the number of outliers in each numeric variable. We can infer that the variable K has the highest number of outliers.

- However, it is not wise to remove those outliers as each point in relevant in determining the type of crops to grow in a certain environment

# Machine Learning Analysis

First we Import the relevant libraries from scikit-learn. 

The reason we do so is because it is easy for us to utilise the machine learning tools and models in the library. Additionally, scikit-learn is useful for our use case as we are using a supervised learning model on a clearly labbelled dataset.

```python
# Library for ML stuff
from sklearn.metrics import classification_report
from sklearn import metrics
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

# To help split into test and trainset much easily
from sklearn.model_selection import train_test_split

# Import Machine learning models
from sklearn.linear_model import LinearRegression


```

Now, we will have to split our data, into the feature (variables which cause change) and target (dependent variable ie what we are trying to guess).

Data Split: 80% train, 20% test

Stratify the split: meaning the 80/20 ratio will be taken per type of crop, rather than the whole dataset,

random_state: Randomise the selection of 80/20. (we chose 32, so we can reproduce results if random state of 32 is chosen again in another function alter)

```python
features = df[['Nitrogen', 'Phosphorus', 'Potassium', 'Temperature', 'Humidity', 'pH',
       'Rainfall']]
target = df['Crop']
labels = df['Crop']

# Split into Test set and Train set. 80% Train, 20% Test.
Xtrain, Xtest, Ytrain, Ytest = train_test_split(features,target,test_size = 0.2, stratify= target, random_state = 32)
```
